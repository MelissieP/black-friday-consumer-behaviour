```python
data["User_ID"].nunique()
new_data = data.groupby(['User_ID','Gender',"Age", "Occupation", "City_Category", "Stay_In_Current_City_Years", 
              "Marital_Status"], as_index=False)['Purchase'].sum()
```

```python
new_data.count()
```

## Gender

```python
sns.set(rc={'figure.figsize':(10,8)})
test = new_data[["Gender", "Age", "Purchase"]].sort_values("Age")
```

```python
test.head()
```

```python
sns.barplot(x = "Age", y = "Purchase", hue = "Gender", data = test)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Across all ages, men spend more money on their purchases than women during black friday. Let's look at married vs unmarried men.

```python
sns.set(rc={'figure.figsize':(10,8)})
test = new_data[["Gender", "Age", "Purchase", "Marital_Status"]].sort_values("Age")

sns.barplot(x = "Age", y = "Purchase", hue = "Marital_Status", data = test)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Seems like unmarried people spend more money during Black Friday, except for the 36-50 age group. This could be due to a family going out and buying products. Let's look at each gender.

```python
test_men = test[test["Gender"] == "M"]
sns.barplot(x = "Age", y = "Purchase", hue = "Marital_Status", data = test_men)
plt.title("Amount Men Spend during Black Friday Based on Their Age and Marital Status")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
test_fem = test[test["Gender"] == "F"]
sns.barplot(x = "Age", y = "Purchase", hue = "Marital_Status", data = test_fem)
plt.title("Amount Women Spend during Black Friday Based on Their Age and Marital Status")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
